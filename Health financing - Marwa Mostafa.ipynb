```python
grouped_reset['Year'].unique()
```

```python
# define x-axis
x1 = grouped_reset[grouped_reset['Region']=='Europe']['Year']
x2 = grouped_reset[grouped_reset['Region']=='Western Pacific']['Year']
x3 = grouped_reset[grouped_reset['Region']=='Americas']['Year']
x4 = grouped_reset[grouped_reset['Region']=='South-East Asia']['Year']
x5 = grouped_reset[grouped_reset['Region']=='Eastern Mediterranean']['Year']
x6 = grouped_reset[grouped_reset['Region']=='Africa']['Year']

# Multiple lines on the same plot
y1 = grouped_reset[grouped_reset['Region']=='Europe']['Value']
y2 = grouped_reset[grouped_reset['Region']=='Western Pacific']['Value']
y3 = grouped_reset[grouped_reset['Region']=='Americas']['Value']
y4 = grouped_reset[grouped_reset['Region']=='South-East Asia']['Value']
y5 = grouped_reset[grouped_reset['Region']=='Eastern Mediterranean']['Value']
y6 = grouped_reset[grouped_reset['Region']=='Africa']['Value']

plt.plot(x1, y1, label='Europe')                # First line
plt.plot(x2, y2, label='Western Pacific')       # Second line
plt.plot(x3, y3, label='Americas')              # Third line
plt.plot(x4, y4, label='South-East Asia')       # Fourth line
plt.plot(x5, y5, label='Eastern Mediterranean') # Fifth line
plt.plot(x6, y6, label='Africa')                # Sixth line

x_labels = grouped_reset['Year'].unique()
plt.xticks(ticks=x_labels, labels=None, rotation=90)

# Adding title and labels
plt.title("Health Expenditure per capita for Regions across years")
plt.xlabel("Year")
plt.ylabel("Health Expenditure Amount per capita in $US")

# Adding a legend to distinguish lines
plt.legend()

# Display the plot
plt.show()
```

```python
df_reg_year = che_per_capita_U6.groupby(['Region','Year'])[['Value']].sum().reset_index().sort_values('Year')

years = df_reg_year.iloc[:,1].unique()
Europe_bar = df_reg_year[df_reg_year['Region']=='Europe']['Value']
Western_Pacific_bar = df_reg_year[df_reg_year['Region']=='Western Pacific']['Value']
Americas_bar = df_reg_year[df_reg_year['Region']=='Americas']['Value']
South_East_Asia_bar = df_reg_year[df_reg_year['Region']=='South-East Asia']['Value']
Eastern_Mediterranean_bar = df_reg_year[df_reg_year['Region']=='Eastern Mediterranean']['Value']
Africa_bar = df_reg_year[df_reg_year['Region']=='Africa']['Value']

subcategories = ['Europe','Western Pacific','Americas','South-East Asia','Eastern Mediterranean','Africa']

width = 0.8
years_pos = np.arange(len(years))

colors = sns.color_palette("muted", len(subcategories))

plt.bar(years_pos, Europe_bar, width, label='Europe')                                                                                            #
plt.bar(years_pos, Western_Pacific_bar, width, bottom=Europe_bar, label='Western Pacific')
plt.bar(years_pos, Americas_bar, width, bottom=np.array(Europe_bar) + np.array(Western_Pacific_bar), label='Americas')
plt.bar(years_pos, South_East_Asia_bar, width, bottom=np.array(Europe_bar) + np.array(Western_Pacific_bar) + np.array(Americas_bar), label='South-East Asia')
plt.bar(years_pos, Eastern_Mediterranean_bar, width, bottom=np.array(Europe_bar) + np.array(Western_Pacific_bar) + np.array(Americas_bar) + np.array(South_East_Asia_bar), label='Eastern Med')
plt.bar(years_pos, Africa_bar, width, bottom=np.array(Europe_bar) + np.array(Western_Pacific_bar) + np.array(Americas_bar) + np.array(South_East_Asia_bar) + np.array(Eastern_Mediterranean_bar), label='Africa')


plt.xticks(years_pos, years, rotation=90)
plt.xlabel('Year')
plt.ylabel('Health Expenditure Value per capita in $US')
plt.title('Cumulative Health Expenditure of Regions across years')
plt.legend()
plt.show()
```
